```python
# Fill simple categorical missing values
df['Occupation'] = df['Occupation'].fillna('Unknown')
df['Exercise_Type'] = df['Exercise_Type'].fillna('Not Specified')
```

```python
# Check if missing values correlate with other variables
print(df[df['Sleep_Duration'].isnull()]['Age'].describe())
print(df[df['Blood_Pressure'].isnull()]['Stress_Detection'].value_counts())
```

```python
# Create age groups
age_bins = [0, 30, 40, 50, 100]
age_labels = ['18-30', '31-40', '41-50', '51+']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

# Check typical sleep duration by age group
print("Average Sleep Duration by Age Group:")
print(df.groupby('Age_Group',observed=True)['Sleep_Duration'].median())
```

```python
# Fill missing sleep duration with median of their age group
df['Sleep_Duration'] = df.groupby('Age_Group',observed=True)['Sleep_Duration'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
# Check typical blood pressure by stress level
print("\nAverage Blood Pressure by Stress Level:")
print(df.groupby('Stress_Detection')['Blood_Pressure'].median())
```

```python
# Fill missing blood pressure with median of their stress level group
df['Blood_Pressure'] = df.groupby('Stress_Detection')['Blood_Pressure'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
# Final check: Remaining missing values
print("\nRemaining Missing Values:")
print(df.isnull().sum())
```

```python
print("Current unique Exercise_Type values:")
print(df['Exercise_Type'].value_counts())
```

```python
df.describe()
```

```python
# Normalize lowercase and strip spaces
df['Exercise_Type'] = df['Exercise_Type'].str.lower().str.strip()

# Mapping for normalized labels
exercise_mapping = {
    'strength training': 'Strength Training',
    'st': 'Strength Training',
    'cardio': 'Cardio',
    'yoga': 'Yoga',
    'yhoga': 'Yoga',
    'pilates': 'Pilates',
    'meditation': 'Meditation',
    'aerobics': 'Aerobics',
    'walking': 'Walking',
    'not specified': 'Not Specified'
}

# Apply mapping
df['Exercise_Type'] = df['Exercise_Type'].replace(exercise_mapping)

# If anything remains unmapped → set to Not Specified
df['Exercise_Type'] = df['Exercise_Type'].fillna('Not Specified')

# Ensure proper Title Case formatting
df['Exercise_Type'] = df['Exercise_Type'].str.title()

# Verify
print("\nUnique Exercise Types after full cleaning:")
print(df['Exercise_Type'].value_counts())
```

# Step 2: Identify Missing or Unusual Values

```python
print("REMAINING MISSING VALUES:")
print(df.isnull().sum())
```

```python
print("POTENTIAL OUTLIERS CHECK (values beyond 3 standard deviations):")
numerical_cols = df.select_dtypes(include=[np.number]).columns

for col in numerical_cols:
    mean = df[col].mean()
    std = df[col].std()
    lower_bound = mean - 3 * std
    upper_bound = mean + 3 * std
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    if len(outliers) > 0:
        print(f"\n{col}: {len(outliers)} potential outliers")
        print(f"   Range: {df[col].min():.2f} to {df[col].max():.2f}")
        print(f"   Outlier bounds: {lower_bound:.2f} to {upper_bound:.2f}")
        print(f"   Outlier values: {outliers[col].unique()}")
    else:
        print(f"{col}: No outliers detected")
```

```python
# Check for unusual values in categorical variables
print("\nUNUSUAL CATEGORICAL VALUES:")
print("Occupation has", df['Occupation'].nunique(), "unique categories - very diverse!")

# Convert times to proper datetime for correct chronological analysis
from datetime import datetime

def convert_to_time(time_str):
    return datetime.strptime(time_str, '%I:%M %p')

# Analyze Wake_Up_Time
wake_times_converted = df['Wake_Up_Time'].apply(convert_to_time)
earliest_wake = wake_times_converted.min()
latest_wake = wake_times_converted.max()

print(f"Wake_Up_Time range: {earliest_wake.strftime('%I:%M %p')} to {latest_wake.strftime('%I:%M %p')}")

# Analyze Bed_Time
bed_times_converted = df['Bed_Time'].apply(convert_to_time)
earliest_bed = bed_times_converted.min()
latest_bed = bed_times_converted.max()

print(f"Bed_Time range: {earliest_bed.strftime('%I:%M %p')} to {latest_bed.strftime('%I:%M %p')}")

# Check for extreme sleep schedules (potential outliers)
print("\nEXTREME SLEEP SCHEDULES:")
print(f"Very early risers (before 5 AM): {len(df[df['Wake_Up_Time'].isin(['4:30 AM'])])}")
print(f"Very late sleepers (after midnight): {len(df[df['Bed_Time'].isin(['12:00 AM', '12:30 AM', '1:00 AM'])])}")
print(f"Very early sleepers (before 6 PM): {len(df[df['Bed_Time'].isin(['4:00 PM', '4:30 PM', '5:00 PM'])])}")
```

# Step 3: Explore Relationships Between Variables

```python
# Let's start with correlation analysis
print("CORRELATION ANALYSIS:")
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlations
correlation_matrix = df.select_dtypes(include=[np.number]).corr()

# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables')
plt.tight_layout()
plt.show()
```

#### A correlation matrix uses a color scale and numerical values to represent the Person correlation coefficient ($r$) between every possible pair of variables.

#### What it Shows:Correlation Coefficient ($r$): The number in each cell, ranging from -1.00 to +1.00, indicates the strength and direction of the linear relationship between the two variables corresponding to that cell's row and column.+1.00: Perfect positive correlation (as one variable increases, the other increases proportionally).-1.00:

#### Perfect negative correlation (as one variable increases, the other decreases proportionally).0.00: No linear correlation.Color Scale (Heatmap): The colors on the right (and in the cells) visually represent the strength and direction.Red/Darker Colors (e.g., Blood-Sugar_Level vs. Cholesterol_Level at 0.94): Strong positive correlation.Blue/Lighter Colors (e.g., Sleep_Duration vs. Travel_Time at -0.31): Strong negative correlation.White/Neutral Colors (e.g., Work_Hours vs. Alcohol_Intake at -0.03): Weak or no correlation.Diagonal: 
#### The main diagonal always has 1.00 because a variable is perfectly correlated with itself.

#### The heatmap reveals a crucial pattern: Stress, Anxiety, and Depression are highly intercorrelated ($r \approx 0.82-0.90$), confirming they often co-occur. Furthermore, these psychological distress factors have a strong positive link with Headache/Physical Symptoms ($r \approx 0.70$), emphasizing a significant mind-body connection. Critically, Social Support shows a moderate-to-strong negative correlation with all negative factors ($r \approx -0.65$ with Stress). This positions social support as a key protective factor, suggesting that strengthening community connections could be an impactful intervention strategy for overall mental well-being in this dataset.

```python
plt.figure(figsize=(14, 10))

# 1․1 Create subplots for a comprehensive sleep overview
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

counts = df['Sleep_Quality'].value_counts()
# Keep only categories > 5%
counts_filtered = counts[counts > 0.05 * counts.sum()]
# Calculate "Other" as the sum of all the rest
other_sum = counts[~counts.index.isin(counts_filtered.index)].sum()
counts_filtered['Other'] = other_sum
counts = counts_filtered
colors = ['#FFADAD', '#FFD6A5', '#FDFFB6', '#CAFFBF', '#9BF6FF']

axes[0,0].pie(
    counts,
    labels=counts.index,
    autopct=lambda p: '{:.1f}%'.format(p) if p > 3 else '',
    colors=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD', '#98D8C8', '#F7DC6F'],
    startangle=45,
    wedgeprops={'width':0.8, 'edgecolor':'w', 'linewidth':2}
)
axes[0,0].set_title('SLEEP QUALITY CRISIS: Population Distribution', fontsize=14, fontweight='bold')


# 1.2 Sleep Duration vs Quality
scatter = axes[0,1].scatter(df['Sleep_Duration'], df['Sleep_Quality'], 
                           c=df['Stress_Detection'].map({'Low':0, 'Medium':1, 'High':2}), 
                           cmap='RdYlGn_r', alpha=0.7, s=60)
axes[0,1].set_xlabel('Sleep Duration (Hours)')
axes[0,1].set_ylabel('Sleep Quality (1-5)')
axes[0,1].set_title(' SLEEP DURATION vs QUALITY: The Critical Connection', fontweight='bold')
plt.colorbar(scatter, ax=axes[0,1], label='Stress Level')

# 1.3 Sleep Quality by Occupation (Top 10 - FILTERED No "Unknown")
occupation_data = df[df['Occupation'] != 'Unknown']
top_occupations = occupation_data['Occupation'].value_counts().head(10).index
occupation_sleep = occupation_data[occupation_data['Occupation'].isin(top_occupations)].groupby('Occupation')['Sleep_Quality'].mean().sort_values()
axes[1,0].barh(range(len(occupation_sleep)), occupation_sleep.values)
axes[1,0].set_yticks(range(len(occupation_sleep)))
axes[1,0].set_yticklabels(occupation_sleep.index)
axes[1,0].set_xlabel('Average Sleep Quality')
axes[1,0].set_title(' OCCUPATION IMPACT: Which Jobs Destroy Sleep?', fontweight='bold')

# 1.4 Sleep Quality by Exercise Type (FILTERED - No "Not Specified")
exercise_data = df[df['Exercise_Type'] != 'Not Specified']
exercise_sleep = exercise_data.groupby('Exercise_Type')['Sleep_Quality'].mean().sort_values(ascending=False)
axes[1,1].bar(exercise_sleep.index, exercise_sleep.values, color=['#2E8B57', '#3CB371', '#90EE90', '#98FB98', '#00FA9A', '#32CD32', '#7CFC00'])
axes[1,1].set_xticks(range(len(exercise_sleep.index)))
axes[1,1].set_xticklabels(exercise_sleep.index, rotation=45, ha='right')

axes[1,1].set_ylabel('Average Sleep Quality')
axes[1,1].set_title(' EXERCISE TYPE: Which Workouts Boost Sleep Quality?', fontweight='bold')

plt.tight_layout()
plt.show()
```

#### Graph 1. DIGITAL ADDICTION: Screen Time vs Sleep Quality (Scatter Plot)
This plot shows a pretty clear pattern: as Screen Time increases, Sleep Quality tends to dip. Most people who report higher screen usage (6+ hours) seem stuck at or below a moderate Sleep Quality of 4.0 or lower. The healthiest sleep scores (above 4.5) are overwhelmingly concentrated in the low-to-medium screen time range (2 to 4 hours).

Crucially, when looking at the stress levels (dot colors), it's visible that while High Stress is present everywhere, the people with the worst possible combination—high Screen Time and the lowest Sleep Quality (2.0)—are predominantly experiencing medium or high stress. It strongly suggests that excessive time in front of screens isn't just sacrificing sleep quality, but is also a likely partner in emotional distress.

#### Graph 2.AGE VS SCREEN TIME: Generational Divide (Box Plot)
The box plots illustrate a distinct generational difference in screen habits. It's not a simple upward or downward trend across all ages, but the data suggests that people in their 30s and 40s (31-40 and 41-50 age groups) are spending the most time on screens, with the highest median screen times and the widest overall range of usage.

The 18-30 year-olds actually show a slightly lower median screen time than their older counterparts. This finding is interesting because it flips the typical stereotype; perhaps this group is balancing screen use with other activities or work, while the middle-aged groups are juggling work, family, and digital leisure, leading to higher sustained screen exposure.

#### Graph 3. ACTIVITY TRADE-OFF: Screen Time vs Physical Activity (Scatter Plot)
Most people fall below the line where screen time and physical activity are equal, which means they usually spend more time on screens than being physically active. A lot of the points are grouped around 3–5 hours of screen time and mid-level physical activity. So it doesn’t look like people are completely inactive, but their lifestyle is definitely more screen-focused.

When you look at sleep quality, a clearer pattern starts to show. People who are more physically active tend to have better sleep, even if their screen time is high. On the other hand, screen time by itself doesn’t seem to tell us much about sleep quality — it’s really the activity level that makes the bigger difference.

Overall, this suggests that encouraging more physical movement might improve sleep more effectively than just trying to cut down screen time alone.

# Visualization 5: The Exercise Prescription - What Really Works?

```python
plt.figure(figsize=(16, 6))

# 5.1 Exercise Type Benefits Matrix - FILTERED: Remove "Not Specified"
plt.subplot(1, 2, 1)
# Filter out "Not Specified" exercise types
exercise_data = df[df['Exercise_Type'] != 'Not Specified']

# Make sure Stress_Numeric exists
if 'Stress_Numeric' not in exercise_data.columns:
    exercise_data['Stress_Numeric'] = exercise_data['Stress_Detection'].map({'Low': 0, 'Medium': 1, 'High': 2})

exercise_benefits = exercise_data.groupby('Exercise_Type')[['Sleep_Quality', 'Physical_Activity', 'Stress_Numeric']].mean()

sns.heatmap(exercise_benefits, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('EXERCISE PRESCRIPTION: Benefits Matrix by Type', fontweight='bold')

# 5.2 Exercise Frequency vs Health Outcomes - FILTERED: Remove "Not Specified"
plt.subplot(1, 2, 2)
# Use the filtered data
sns.scatterplot(data=exercise_data, x='Physical_Activity', y='Sleep_Quality', 
                hue='Exercise_Type', alpha=0.7, s=80, palette='viridis')

# Add trend line to show overall relationship
import numpy as np
z = np.polyfit(exercise_data['Physical_Activity'], exercise_data['Sleep_Quality'], 1)
p = np.poly1d(z)
plt.plot(exercise_data['Physical_Activity'], p(exercise_data['Physical_Activity']), "r--", alpha=0.8, linewidth=2)

plt.title('EXERCISE FREQUENCY: Optimal Doses for Health', fontweight='bold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

# Visualization 8: Health Metrics Correlation Network

```python
plt.figure(figsize=(12, 10))

# Comprehensive Health Correlation - FIXED: Use numeric stress
plt.subplot(1, 1, 1)

# Make sure Stress_Numeric exists
if 'Stress_Numeric' not in df.columns:
    df['Stress_Numeric'] = df['Stress_Detection'].map({'Low': 0, 'Medium': 1, 'High': 2})

health_corr = df[['Sleep_Quality', 'Blood_Pressure', 'Cholesterol_Level', 
                  'Blood_Sugar_Level', 'Physical_Activity', 'Stress_Numeric']].corr()

mask = np.triu(np.ones_like(health_corr, dtype=bool))
sns.heatmap(health_corr, mask=mask, annot=True, cmap='RdYlBu_r', center=0, 
            square=True, fmt='.2f')
plt.title('HEALTH METRICS NETWORK: Interconnected System', fontweight='bold', fontsize=14)
plt.tight_layout()
plt.show()

print("STRONGEST HEALTH CORRELATIONS:")
print("=" * 40)
corr_pairs = health_corr.unstack().sort_values(ascending=False)

# Remove self-correlations and duplicates
corr_pairs = corr_pairs[corr_pairs < 0.999]
strong_correlations = corr_pairs.head(6)

for pair, value in strong_correlations.items():
    var1, var2 = pair
    if var1 != var2:
        print(f"{var1} ↔ {var2}: {value:.3f}")
```
